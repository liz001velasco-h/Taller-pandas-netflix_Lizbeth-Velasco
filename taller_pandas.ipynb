```python
#1. Importar librerías

# Importamos la librería pandas, que se usa para manipulación y análisis de datos
import pandas as pd
```

```python
#2. Cargar el dataset

# Cargamos el archivo CSV que contiene el dataset de Netflix
df = pd.read_csv("netflix_titles.csv")

# Mostramos las primeras 5 filas del dataset para conocer su estructura
df.head()
```

```python
#3. Exploración inicial de los datos

# Ver la cantidad de filas y columnas del dataset
df.shape

# Mostrar los nombres de todas las columnas del dataset
df.columns

# Obtener información general del dataset:
# tipo de datos, cantidad de valores no nulos y memoria utilizada
df.info()

# Mostrar estadísticas descriptivas de las columnas
# include="all" permite analizar también columnas de texto
df.describe(include="all")
```

```python
#4. Identificación de valores nulos

# Identificar cuántos valores nulos existen en cada columna
# Esto es importante para el proceso de limpieza de datos
df.isnull().sum()
```

```python
#5. Limpieza de datos

# Reemplazar valores nulos en la columna "director"
# Se reemplazan por la palabra "Desconocido"
df["director"] = df["director"].fillna("Desconocido")

# Reemplazar valores nulos en la columna "country"
df["country"] = df["country"].fillna("No especificado")

# Eliminar filas donde no exista información en la columna date_added, ya que esta variable se utilizará para análisis posteriores
df = df.dropna(subset=["date_added"])
```

```python
#6. Transformación de datos

# Algunos registros tienen espacios al inicio de la fecha, lo que genera errores al convertir el formato.
# Por eso primero eliminamos los espacios con str.strip()
# errors="coerce" hace que si alguna fecha está mal escrita se convierta en NaN en lugar de generar error.

df["date_added"] = df["date_added"].str.strip()

df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
```

```python
#7. Creación de nuevas variables

# Crear una nueva columna llamada "year_added"
# Esta columna contiene el año en que el contenido fue agregado a Netflix
df["year_added"] = df["date_added"].dt.year

# Visualizar nuevamente el dataset para confirmar la nueva columna creada
df.head()
```

```python
#8. Selección de columnas

# Seleccionar solo algunas columnas del dataset
# Esto permite trabajar únicamente con la información necesaria
df[["title", "type", "country", "release_year"]].head()
```

```python
#9. Aplicar filtros

#Películas después de 2015

# Filtrar únicamente las películas cuyo año de lanzamiento sea mayor a 2015
movies_2015 = df[(df["type"] == "Movie") & (df["release_year"] > 2015)]

# Mostrar las primeras filas del resultado
movies_2015.head()
```

```python
#Series producidas en Estados Unidos

# Filtrar únicamente las series de televisión producidas en Estados Unidos
series_usa = df[(df["type"] == "TV Show") & (df["country"] == "United States")]

# Mostrar algunas filas del resultado
series_usa.head()
```

```python
#10. Agrupar datos

#Cantidad de contenidos por tipo

# Agrupar los datos por tipo de contenido (Movie o TV Show), y contar la cantidad total de títulos en cada categoría
df.groupby("type")["title"].count()
```

```python
#Cantidad de contenidos por país

# Agrupar los datos por país y contar cuántos títulos existen por cada uno
# Luego ordenar los resultados de mayor a menor
df.groupby("country")["title"].count().sort_values(ascending=False).head(10)
```

```python
#Contenidos agregados por año

# Agrupar los datos según el año en que fueron agregados a Netflix
df.groupby("year_added")["title"].count()
```

```python
#11. Crear nuevas variables

# Crear una nueva columna que clasifique el contenido como "Reciente" o "Antiguo"
# Si el año de lanzamiento es mayor o igual a 2015 se considera reciente
df["categoria_estreno"] = df["release_year"].apply(
    lambda x: "Reciente" if x >= 2015 else "Antiguo"
)

# Visualizar el dataset con la nueva variable creada
df.head()
```

```python
#12. Análisis de datos

#Preguntas de análisis

#Pregunta 1

#¿Qué tipo de contenido predomina en Netflix: películas o series?

# Contar cuántos registros existen para cada tipo de contenido
df["type"].value_counts()
```

```python
# Conclusión del análisis para la pregunta 1

print("CONCLUSIÓN 1: El análisis muestra que las películas representan la mayor parte del contenido disponible en Netflix, lo que indica que la plataforma tiene una mayor oferta de películas en comparación con las series de televisión.")
```

```python
#Pregunta 2

#¿Qué países producen más contenido para Netflix?

# Identificar los países con mayor cantidad de títulos en la plataforma
df["country"].value_counts().head(10)
```

```python
#Conclusión del análisis para la pregunta 2

print("CONCLUSIÓN 2: Estados Unidos lidera la producción de contenido dentro del catálogo de Netflix, seguido por países como India y Reino Unido, lo que refleja la fuerte presencia de la industria audiovisual estadounidense en la plataforma.")
```

```python
#Pregunta 3

#¿En qué años se agregaron más contenidos a Netflix?

# Analizar cuántos contenidos fueron agregados por año
df["year_added"].value_counts().sort_index()
```

```python
#Conclusión del análisis para la pregunta 3

print("CONCLUSIÓN 3: A partir de los resultados se observa que desde 2016 la cantidad de contenidos agregados a Netflix aumentó significativamente, lo que coincide con el crecimiento global de la plataforma y su expansión en diferentes países.")
```

```python
#13. Guardar el dataset limpio

# Guardar el dataset después de la limpieza y transformación
# index=False evita que se guarde la columna de índices
df.to_csv("netflix_limpio.csv", index=False)
```
